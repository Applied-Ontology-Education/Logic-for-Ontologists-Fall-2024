```python
def runRobot(model="", reasoner = "HERMIT", output="", error=""):
    cmdline = f"robot reason --reasoner {reasoner} --input {model} --output {output} > {error}"
    print(cmdline)
    res = os.system(cmdline)
    return res
    
    
    
```

```python
model_types = [
    "all_different_individuals"
    , "all_same_individuals"
    , "no_interindividual_assertion"
]
reasoners = ['HERMIT','ELK']
results= {}

for model_type in model_types:
    print(model_type)
    for reasoner in reasoners:

        if model_type not in results: 
            results[model_type] = {}   
            
        if reasoner not in results[model_type]: 
            results[model_type][reasoner] = {}
            
        
        input_path = f"{model_type}/"
        log_path= f"{model_type}_logs/"
        output_path  = f"{model_type}_output/"

        for fn in os.listdir(input_path):
            
            if not fn[-4:] == '.ttl': 
                print(fn, ' miss') 
                continue
            
            input_file = f"{input_path}{fn}"
            output = f"{output_path}{fn}"
            error = f'{log_path}{reasoner.lower()}_{fn.replace(".ttl",".txt")}'

            runRobot(model=input_file, reasoner=reasoner, output=output, error=error)
            try:
                sz = os.path.getsize(error)
            except:
                results[model_type][reasoner][fn] = {"log_file_present": False}
                
            
            if fn not in results[model_type][reasoner]:
                results[model_type][reasoner][fn] = {"error_file_size": sz,"log_file_present": True }

import json
open('consistency_check_results.json','w').write(json.dumps(results, indent=2)) 

```
